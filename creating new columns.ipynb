Graph of inner Graph

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.random.randint(1,10,5)
y=x**2
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.3,0.3,0.4,0.4])

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

Subpolts

```python
fig=plt.figure()
fig,axes=plt.subplots(2,3)

plt.subplot(2,3,1)
plt.plot(x,y,'r*')

plt.subplot(2,3,2)
plt.plot(y,x,'y-')

plt.subplot(2,3,4)
plt.plot(x,y,'c-')
```

maximum properties in a graph

```python
fig=plt.figure()
a=fig.add_axes([0,0,1,1])
a.plot(x,y,'g',lw=1,ls='-',marker='o',markersize=20,markerfacecolor='y',markeredgewidth=2, markeredgecolor='c')
```
